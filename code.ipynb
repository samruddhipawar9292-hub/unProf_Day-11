```python
from textblob import TextBlob
```

```python
review = "This dress is amazing!"

blob = TextBlob(review)

blob.sentiment
```

```python
def get_sentiment(review):

    polarity = TextBlob(review).sentiment.polarity

    if polarity > 0:
        return "Positive"

    elif polarity < 0:
        return "Negative"

    else:
        return "Neutral"
```

```python
def get_polarity(review):
    return TextBlob(review).sentiment.polarity
```

```python
df_sample["Polarity"] = df_sample["Review Text"].apply(get_polarity)
```

```python
df_sample["Sentiment"] = df_sample["Review Text"].apply(get_sentiment)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
review = "This dress is amazing! I absolutely love it."

score = analyzer.polarity_scores(review)

score
```

```python
def get_vader_score(review):
    return analyzer.polarity_scores(review)["compound"]
```

```python
def get_vader_sentiment(review):

    compound = analyzer.polarity_scores(review)["compound"]

    if compound >= 0.05:
        return "Positive"

    elif compound <= -0.05:
        return "Negative"

    else:
        return "Neutral"
```

```python
df_sample["VADER Score"] = df_sample["Review Text"].apply(get_vader_score)
```

```python
df_sample["VADER Sentiment"] = df_sample["Review Text"].apply(get_vader_sentiment)
```
